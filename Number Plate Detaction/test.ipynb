```python
from ultralytics import YOLO
import torch
from pathlib import Path

assert torch.cuda.is_available()

model = YOLO("Number Plate Detaction/license_plate_detector.pt").to("cuda")

video_dir = Path("D:\\Luminar\\Study\\Deep learning\\DL Project\Yolo\\Number Plate Detaction")
video_exts = {".mp4", ".avi", ".mov", ".mkv"}

for video in video_dir.iterdir():
    if video.suffix.lower() in video_exts:
        print(f"Processing: {video.name}")
        model.predict(
            source=str(video),
            device=0,
            save=True,
            conf=0.25,
            imgsz=640
        )
```
