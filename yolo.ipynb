```python
import cv2
import ultralytics
```

```python
model=ultralytics.YOLO("yolov8n.pt")
video = cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img=cv2.flip(img,1)
    result=model([img])
    cv2.imshow("webcam",result[0].plot())
    if cv2.waitKey(1) & 0xff==ord("q"):
        break

video.release()
cv2.destroyAllWindows()
```
